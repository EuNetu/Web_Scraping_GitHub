```python
import pandas as pd
import numpy as np

import requests as rq
from bs4 import BeautifulSoup
import re
from google.colab import drive

drive.mount('/content/drive')

%cd /content/drive/MyDrive/ColabNotebooks

lista = ['https://github.com/nadinibzrra',
         'https://github.com/EuNetu',
         'https://github.com/lauraemmanuella',
         'https://github.com/taniro',
         'https://github.com/GeorgeOgeorge',
         'https://github.com/Thiazex',
         'https://github.com/dalania',
         'https://github.com/manelB1',
         'https://github.com/Enuch',
         'https://github.com/mariorafael21',
         'https://github.com/jean779',
         'https://github.com/Gus-1003']


info_dados = []
info_dados_final = []
info_dados = list()

for i in lista:
  dados = {}
  try:
    url = i
    response = rq.get(url)
    pagina_html = BeautifulSoup(response.text,'html.parser') #Cria objeto BeautifulSoup com analisador de HTML (parser)

    tag_h1 = pagina_html.find('h1', {'class': 'vcard-names'})
    nome = tag_h1.findAll('span')
    dados['nome'] = nome[0].get_text()
    dados['username'] = nome[1].get_text()

    tag_span = pagina_html.findAll('span', {'class': 'text-bold color-fg-default'})
    dados['seguidores'] = tag_span[0].get_text()
    dados['seguindo'] = tag_span[1].get_text()
    dados['favoritos'] = tag_span[2].get_text()

    tag_a = pagina_html.find('a', {'class': 'UnderlineNav-item js-responsive-underlinenav-item'})
    repositorioValor = tag_a.findAll('span')
    dados['numero de repositórios'] = repositorioValor[0].get_text()

    repositorio_popular = pagina_html.find('span', {'class': 'repo'})
    dados['primeiro repositorio pinado'] = repositorio_popular.get_text()

    linguagem = pagina_html.find('span', {'itemprop': 'programmingLanguage'})
    dados['linguagem do primeiro repositório pinado'] = linguagem.get_text()

    info_dados.append(dados)
    
    
  except Exception as e:
    print(e)


df_info = pd.DataFrame(info_dados)
df_info.to_csv("banco_de_dados.csv")
df_info


```
